```python
# Initialise LLM with required params
llm = OpenAI(temperature=0.9, max_tokens=500) 
```

```python
loaders = UnstructuredURLLoader(urls=[
    "https://www.moneycontrol.com/news/business/markets/wall-street-rises-as-tesla-soars-on-ai-optimism-11351111.html",
    "https://www.moneycontrol.com/news/business/tata-motors-launches-punch-icng-price-starts-at-rs-7-1-lakh-11098751.html"
])
data = loaders.load() 
len(data)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

# As data is of type documents we can directly use split_documents over split_text in order to get the chunks.
docs = text_splitter.split_documents(data)
```

```python
# Ensure that docs is a list of strings
docs = [str(doc) for doc in docs]
```

```python
import time
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import FAISS

# Initialize OpenAIEmbeddings
embeddings = OpenAIEmbeddings()

# Initialize a list to store embeddings for each document
embedded_docs = []

# Process each document with error handling and delay to avoid rate limits
for doc in docs:
    try:
        # Create embeddings for the document (use list format as expected by embed_documents)
        doc_embedding = embeddings.embed_documents([doc])
        
        # Check that the embedding is not empty
        if doc_embedding:
            embedded_docs.extend(doc_embedding)  # Use extend to get embeddings directly in the list

        # Add a delay to avoid rate limits
        time.sleep(1)
    except Exception as e:
        print(f"Error creating embedding for document {doc}: {e}")

# Verify that we have embeddings before creating the FAISS index
if embedded_docs:
    # Pass to FAISS
    vectorindex_openai = FAISS.from_texts(docs, embeddings)
else:
    print("No embeddings created. Please check your documents and embedding function.")
```

```python
from sentence_transformers import SentenceTransformer
from langchain.vectorstores import FAISS
from langchain.embeddings.base import Embeddings

# Load the local embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')  # Choose any Sentence Transformer model

# Define a wrapper class for the Sentence Transformer model
class SentenceTransformerEmbeddings(Embeddings):
    def embed_documents(self, texts):
        return [model.encode(text) for text in texts]
    
    def embed_query(self, text):
        return model.encode(text)

# Instantiate the custom embedding wrapper
sentence_embeddings = SentenceTransformerEmbeddings()

# Create the FAISS index
vectorindex = FAISS.from_texts(docs, embedding=sentence_embeddings)
```

```python
import time
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import FAISS

# Initialize OpenAIEmbeddings
embeddings = OpenAIEmbeddings()

# List to hold individual document embeddings
embedded_docs = []

# Process each document individually to avoid rate limit errors
for doc in docs:
    try:
        # Get embeddings for the current document
        embedded_docs.append(embeddings.embed_documents([doc]))
        
        # Delay to prevent rate limiting
        time.sleep(1)  # Adjust the delay as needed
    except Exception as e:
        print(f"Error creating embedding for document {doc}: {e}")

# Create FAISS vector index from embedded documents
vectorindex_openai = FAISS.from_documents(embedded_docs, embeddings)

```
